```python
# IMPORTANT: RUN THIS CELL IN ORDER TO IMPORT YOUR KAGGLE DATA SOURCES,
# THEN FEEL FREE TO DELETE THIS CELL.
# NOTE: THIS NOTEBOOK ENVIRONMENT DIFFERS FROM KAGGLE'S PYTHON
# ENVIRONMENT SO THERE MAY BE MISSING LIBRARIES USED BY YOUR
# NOTEBOOK.
import kagglehub
ismailnasri20_driver_drowsiness_dataset_ddd_path = kagglehub.dataset_download('ismailnasri20/driver-drowsiness-dataset-ddd')

print('Data source import complete.')
```

```python
import os

base_path = "/root/.cache/kagglehub/datasets/ismailnasri20/driver-drowsiness-dataset-ddd/versions/1/Driver Drowsiness Dataset (DDD)"

drowsy_path = os.path.join(base_path, "Drowsy")
alert_path  = os.path.join(base_path, "Non Drowsy")

def count_images(path):
    return len([f for f in os.listdir(path) if f.lower().endswith(('.jpg','.png','.jpeg'))])

drowsy_count = count_images(drowsy_path)
alert_count  = count_images(alert_path)

print("Jumlah Drowsy :", drowsy_count)
print("Jumlah NonDrowsy :", alert_count)

# cek balance
if drowsy_count > 0 and alert_count > 0:
    ratio = max(drowsy_count, alert_count) / min(drowsy_count, alert_count)
    print(f"Rasio: {ratio:.2f}")
    if ratio > 1.5:
        print("Dataset tidak seimbang.")
    else:
        print("Dataset cukup seimbang.")
else:
    print("Salah satu kategori kosong.")
```
